```python
import numpy as np
from itertools import product

tiles = np.array([list(i) for i in input.split('\n')])

# Get symbol indices
symbol_indices = []
for i, val in np.ndenumerate(tiles):
    if val == '*':
        symbol_indices.append(i)

# for each symbol index, walk the surroudings
sum_of_prods = 0
for symbol_index in symbol_indices:
    word_set = set()
    
    for i_offset in range(-1,2,1):
        for j_offset in range(-1,2,1):
            i_prime = min( max(0, symbol_index[0] + i_offset), tiles.shape[0])
            j_prime = min( max(0, symbol_index[1] + j_offset), tiles.shape[1])
            
            if tiles[i_prime, j_prime].isnumeric():
                j_start = j_prime
                j_end = j_prime
                
                more_to_find = True
                while more_to_find:
                    more_to_find = False
                    if j_start-1 >=0 and  tiles[i_prime, j_start-1].isnumeric():
                        j_start -= 1
                        more_to_find = True
                    if  j_end+1 < tiles.shape[1] and tiles[i_prime, j_end+1].isnumeric():
                        j_end += 1
                        more_to_find = True
                word_set.add( tuple([tuple([i_prime, j_start])
                                                        , tuple([i_prime, j_end])
                                                       ]
                                                      )
                                               )
    
    if len(word_set)>1:
        prod = 1
        for word_ind in word_set:
            word = int(''.join( tiles[word_ind[0][0], word_ind[0][1]:word_ind[1][1]+1]))
            prod *= word        
    else:
        prod = 0
    
    sum_of_prods += prod

print( sum_of_prods )
```
